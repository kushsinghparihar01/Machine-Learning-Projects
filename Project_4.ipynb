```python
import pandas as pd
```

```python
df = pd.DataFrame(i.data,columns = i.feature_names)
df["Species"]=i.target
```

```python
df
```

```python
# GROUPING SETOSA, VERSICOLOR, VERGINA, ON THE BASIS OF SPECIES(0,1,2)
```

```python
s= df.groupby("Species")
```

```python
setosa = s.get_group(0)
versi = s.get_group(1)
virginica = s.get_group(2)
```

```python
setosa.head()
```

```python
versi.head()
```

```python
virginica.head()
```

SPLITTING DATA INTO DEPENDENT AND INDEPENDENT VARIABLE

```python
x = df[["sepal length (cm)","sepal width (cm)","petal length (cm)", "petal width (cm)"]]
```

```python
x.head()
```

```python
y = df["Species"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(setosa["sepal length (cm)"], setosa["sepal width (cm)"],color='red',label="setosa")
plt.scatter(versi["sepal length (cm)"], versi["sepal width (cm)"],color='blue',label="versi",alpha=0.4)
plt.scatter(virginica["sepal length (cm)"], virginica["sepal width (cm)"],color='green',label="virginica")
plt.legend()
```

```python
plt.scatter(setosa["petal length (cm)"], setosa["petal width (cm)"],color='red',label="setosa")
plt.scatter(versi["petal length (cm)"], versi["petal width (cm)"],color='blue',label="versi",alpha=1)
plt.scatter(virginica["petal length (cm)"], virginica["petal width (cm)"],color='green',label="virginica")
plt.legend()
```

```python
y
```

NOW WE TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x, train_y, test_y = train_test_split(x,y,test_size = 0.3)
```

```python
len(train_x)
```

```python
len(test_x)
```

IMPORTING KNN LIBRARY FROM SKLEARN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
yy = model.predict(test_x)==test_y
```
