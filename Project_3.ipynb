```python
import pandas as pd
```

```python
data = pd.read_csv("insurance.csv")
```

```python
data.head()
```

TAKING DATA INTO DEPENDENT("Y") OR IN INDEPENDENT VARIABLE("X")

```python
x = data[["age"]]
```

```python
y = data[["bought_insurance"]]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,c="red")
plt.title("Insurance-chart")
plt.xlabel("Age")
plt.ylabel("Insurance (given or not)")
plt.show()
```

IMPORT LOGISTIC REGRESSION FROM  SK_LEARN

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
model = LogisticRegression()
```

```python
model.fit(x,y)
```

```python
a=int(input("Enter the age of customer "))
pred = model.predict([[a]])[0]
if pred==0:
    print("No")
else:
    print("Yes")
```
