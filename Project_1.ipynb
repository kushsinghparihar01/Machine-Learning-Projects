```python
import pandas as pd
```

```python
df = pd.read_csv("areaprice.csv")
```

```python
df
```

```python
x= df[["Area"]]
```

```python
x
```

```python
y=df[["Price"]]
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area-Price prediction graph")
plt.show()
plt.savefig("My graph1")
```

INSTALLING SCIKIT LEARN

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(pd.DataFrame([[5000]],columns=x.columns))
```

```python
model.coef_    # this is coefficient "M"
```

```python
model.intercept_  # this is intercept "C"
```

```python
model.coef_*5000+model.intercept_  # we manually calculate using y = mx+c both predicted and calculated values are same
```

```python
pred = model.predict(x)  # this gives the predicted value for all x given in actual dataP
```

```python
pred
```

```python
act = df[["Price"]]
```

```python
act
```

```python
# ABOVE YOU CAM SEEE THERE IS NOT MUCH DIFFERENCE BETWEEN ACTUAL VALUE AND PREDICTED VALUE
```

```python
plt.plot(x,pred,label="predicted-value")
plt.scatter(x,act,label="Actual-value",c="red")
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
plt.show()
plt.savefig("My graph2")
```

```python
# ABOVE YOU CAN VISUALIZE THAT THERE IS NOT MUCH DIFFERENCE IN ACTUAL AND PREDICTED DATA
```
